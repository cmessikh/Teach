## How to print
### Print an expression
Write `print("Salam")` in the code cell below:

```python
print("Salam")
```

### Print a value
Assign 11 to a variable x, and try to print x using `print(x)` in the cell below:

```python
x = 11
print(x)
```

### Print a value with a message
We can write multiple items with the same print, putting them as arguments. Try to write the expression "the result is" and the value of that previous x in one `print`

```python
print("the result is", x)
```

We can also form a string by composing items, and then print it, or use it in different problems. Try to print a string s, that holds the same previously printed expression, assigned by using a string formatting: `s = f"the result is {x}"`.

```python
s = f"the result is {x}"
print(s)
```

## How to input

### Simple Input to a Variable
We can assign an input received from a user as a string.

```python
s = input()
print(s)
```

We can add an information message so that the user knows what he is about to input.

```python
s = input("Enter your name:")
print(f"Hello {s}!!!")
```

Do not forget! the input always returns a string!

```python
a = input()
b = input()
c = a + b
print(f"{a} + {b} = {c}")
```

```python
# With the same code
a = input()
b = input()
c = a + b
print(f"{a} + {b} = {c}")
```

### Input a non-string variable
In order to, see the input as an integer or float, we need to force it with int() or float()

```python
a = int(input())
b = input()
c = a + float(b)
print(f"{a} + {b} = {c}")
```
